```python
import pandas as pd
import matplotlib.pyplot as plt
path = "../OriginalData/"
filename = "wielkopolskie.csv"
data = pd.read_csv(path + filename)

formated_data = data.rename(columns={"Unnamed: 0": "Client_no",
                                     "Dni od zakupu": "Days_from_purchase",
                                     "Wiek kupującego": "Buyer_age",
                                     "Płeć kupującego": "Gender",
                                     "Ocena": "Rating",
                                     "Marka": "Brand"})
formated_data
formated_data.to_csv('../AnalysisData/formated_data.csv', index = False)
```

```python
data_sort_by_days_from_purchase = formated_data.sort_values(['Days_from_purchase'], ascending = True)

data_sort_by_days_from_purchase = data_sort_by_days_from_purchase.dropna(subset=['Days_from_purchase'])
max_days = max(data_sort_by_days_from_purchase["Days_from_purchase"])
min_days = min(data_sort_by_days_from_purchase["Days_from_purchase"])
print(max_days)
print(min_days)

days = data_sort_by_days_from_purchase.Days_from_purchase.unique()
    
days_amount = []

for day in days:
    days_amount.append(len(data_sort_by_days_from_purchase.loc[(data_sort_by_days_from_purchase['Days_from_purchase'] == day)]))
plt.figure(figsize = (15, 5))
plt.bar(days, days_amount, width = 0.85)
plt.xlabel("Days from purchase")
plt.ylabel("Quantity")
plt.title("Days_from_purchase for " + filename)
plt.grid(linestyle = ':', linewidth = 0.8, alpha = 0.5)
plt.show()
data_sort_by_days_from_purchase
data_sort_by_days_from_purchase.to_csv('../AnalysisData/data_sort_by_days_from_purchase.csv', index = False)
```

```python
brands = formated_data.Brand.unique()
    
brands_amount = []

for brand in brands: 
    brands_amount.append(len(formated_data.loc[(formated_data['Brand'] == brand)]))
    
plt.figure(figsize = (15, 5))
plt.bar(brands, brands_amount, width = 0.85, color = "brown")
plt.xlabel("Brand of the product")
plt.ylabel("Quantity")
plt.title("Brand for " + filename)
plt.grid(linestyle = ':', linewidth = 0.8, alpha = 0.5)
plt.show()

print(len(brands_amount))
max_qty = max(brands_amount)
min_qty = min(brands_amount)
print(max_qty)
print(min_qty)
```

```python
data_sort_by_age = formated_data.sort_values(['Buyer_age'], ascending = True)
data_sort_by_age = data_sort_by_age.dropna(subset=['Buyer_age'])
max_age = max(data_sort_by_age['Buyer_age'])
min_age = min(data_sort_by_age['Buyer_age'])
print(max_age)
print(min_age)

ages = data_sort_by_age.Buyer_age.unique()
    
ages_amount = []

for age in ages: 
    ages_amount.append(len(data_sort_by_age.loc[(data_sort_by_age["Buyer_age"] == age)]))
plt.figure(figsize = (15, 5))
plt.bar(ages, ages_amount, width = 0.85, color = "lime")
plt.xlabel("Reported age")
plt.ylabel("Quantity")
plt.title("Buyer_age for " + filename)
plt.grid(linestyle = ':', linewidth = 0.8, alpha = 0.5)
plt.show()

data_sort_by_age
data_sort_by_age.to_csv('../AnalysisData/data_sort_by_age.csv', index = False)
```

```python
genders = formated_data.Gender.unique()
    
genders_amount = []

for gender in genders:
    genders_amount.append(len(formated_data.loc[(formated_data['Gender'] == gender)]))
plt.figure(figsize = (15, 5))
plt.bar(genders, genders_amount, width = 0.85, color = "darkblue")
plt.xlabel("Reported gender")
plt.ylabel("Quantity")
plt.title("Days from purchase in " + filename)
plt.grid(linestyle = ':', linewidth = 0.8, alpha = 0.5)
plt.show()

max_g = max(genders_amount)
min_g = min(genders_amount)
print(max_g)
print(min_g)
formated_data
```

```python
data_sort_by_rating = formated_data.sort_values(['Rating'], ascending = True)
data_sort_by_rating = data_sort_by_rating.dropna(subset=['Rating'])

grades = data_sort_by_rating.Rating.unique()
    
grades_amount = []

for grade in grades:
    grades_amount.append(len(data_sort_by_rating.loc[(data_sort_by_rating['Rating'] == grade)]))
plt.figure(figsize = (15, 5))
plt.bar(grades, grades_amount, width = 0.35, color = "orange")
plt.xlabel("Rating for the product")
plt.ylabel("Quantity")
plt.title("Rating for " + filename)
plt.gca().invert_xaxis()
plt.grid(linestyle = ':', linewidth = 0.8, alpha = 0.5)
plt.show()

max_rated = max(grades_amount)
min_rated = min(grades_amount)
print(max_rated)
print(min_rated)

data_sort_by_rating
data_sort_by_rating.to_csv('../AnalysisData/data_sort_by_rating.csv', index = False)
```
